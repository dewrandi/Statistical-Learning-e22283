# Kalman Filter Assignment — Solutions

# Q. Analytical Derivation

## 1. Distribution of $x_k^-$

We are given
$$x^{-}_k = A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \qquad x^{+}_{k-1}\sim\mathscr N(m_{k-1},P_{k-1}), \qquad w_{k-1}\sim\mathscr N(0,\Sigma_p),$$
with $x^+_{k-1}$ and $w_{k-1}$ independent.

$x_k^-$ is an **affine transformation of jointly Gaussian random vectors**, hence it is itself Gaussian. We only need its mean and covariance.

**Mean:**
$$
\mathbb E[x_k^-] = A_{k-1}\,\mathbb E[x_{k-1}^+] + G_{k-1}\,\mathbb E[w_{k-1}] = A_{k-1}m_{k-1} + G_{k-1}\cdot 0 = A_{k-1}m_{k-1} \triangleq m_k^-.
$$

**Covariance:** Let $\tilde x = x_{k-1}^+ - m_{k-1}$. Then
$$
x_k^- - m_k^- = A_{k-1}\tilde x + G_{k-1}w_{k-1}.
$$
Since $\tilde x$ and $w_{k-1}$ are independent (zero-mean),
$$
P_k^- = \mathbb E\big[(x_k^- - m_k^-)(x_k^- - m_k^-)^T\big]
= A_{k-1}\,\mathbb E[\tilde x\tilde x^T]\,A_{k-1}^T + G_{k-1}\,\mathbb E[w_{k-1}w_{k-1}^T]\,G_{k-1}^T
$$
(the cross terms vanish by independence and zero mean), so
$$
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
$$
Therefore $x_k^- \sim \mathscr N(m_k^-, P_k^-)$. $\blacksquare$

## 2. Distribution of $y_k^-$

Given $y_k^- = H_k x_k^- + z_k$ with $z_k\sim\mathscr N(0,\Sigma_m)$ independent of $x_k^-$, $y_k^-$ is again an affine map of jointly Gaussian variables, hence Gaussian.

**Mean:**
$$
\mathbb E[y_k^-] = H_k\,\mathbb E[x_k^-] + \mathbb E[z_k] = H_k m_k^-.
$$

**Covariance:** with $x_k^- - m_k^- $ independent of $z_k$,
$$
\mathrm{Cov}(y_k^-) = H_k\,\mathrm{Cov}(x_k^-)\,H_k^T + \mathrm{Cov}(z_k) = H_k P_k^- H_k^T + \Sigma_m.
$$

Hence
$$
y_k^- \sim \mathscr N\big(H_k m_k^-,\; H_k P_k^- H_k^T + \Sigma_m\big). \qquad \blacksquare
$$

## 3. Joint distribution of $(x_k^-, y_k^-)$

Both $x_k^-$ and $y_k^-$ are affine functions of the jointly Gaussian pair $(x_{k-1}^+, w_{k-1}, z_k)$, so the stacked vector $\begin{bmatrix}x_k^-\\ y_k^-\end{bmatrix}$ is jointly Gaussian. The mean vector is just the stack of the two means found above:
$$
\mathbb E\begin{bmatrix}x_k^-\\ y_k^-\end{bmatrix} = \begin{bmatrix}m_k^-\\ H_k m_k^-\end{bmatrix}.
$$

For the covariance block we already have the diagonal blocks $P_k^-$ and $H_kP_k^-H_k^T+\Sigma_m$. The cross-covariance is, using $y_k^- - H_km_k^- = H_k(x_k^--m_k^-) + z_k$ and independence of $z_k$ from $x_k^-$:
$$
\mathrm{Cov}(x_k^-,y_k^-) = \mathbb E\big[(x_k^--m_k^-)\,(H_k(x_k^--m_k^-)+z_k)^T\big] = P_k^- H_k^T.
$$
By symmetry $\mathrm{Cov}(y_k^-,x_k^-) = H_kP_k^-$. Therefore
$$
\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\!\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right). \qquad \blacksquare
$$

## 4. Conditional (posterior) distribution $x_k^+ = (x_k^- \mid y_k^- = y_k^{\mathrm{obs}})$

For a jointly Gaussian pair $\begin{bmatrix}a\\b\end{bmatrix}\sim\mathscr N\left(\begin{bmatrix}\mu_a\\\mu_b\end{bmatrix},\begin{bmatrix}\Sigma_{aa}&\Sigma_{ab}\\\Sigma_{ba}&\Sigma_{bb}\end{bmatrix}\right)$, the standard Gaussian conditioning formula gives
$$
a\mid b=b^{\rm obs} \sim \mathscr N\big(\mu_a + \Sigma_{ab}\Sigma_{bb}^{-1}(b^{\rm obs}-\mu_b),\; \Sigma_{aa}-\Sigma_{ab}\Sigma_{bb}^{-1}\Sigma_{ba}\big).
$$

Applying this with $a = x_k^-$, $b = y_k^-$, and the joint distribution from Part 3:

- $\mu_a = m_k^-$, $\mu_b = H_km_k^-$
- $\Sigma_{aa} = P_k^-$, $\Sigma_{ab} = P_k^-H_k^T$, $\Sigma_{bb} = H_kP_k^-H_k^T+\Sigma_m$, $\Sigma_{ba}=H_kP_k^-$

Define
$$
K_k \triangleq \Sigma_{ab}\Sigma_{bb}^{-1} = P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}.
$$

**Updated mean:**
$$
m_k = m_k^- + K_k\,(y_k^{\rm obs} - H_km_k^-).
$$

**Updated covariance:**
$$
P_k = P_k^- - K_kH_kP_k^- = (I-K_kH_k)P_k^-.
$$

Hence $x_k^+ \sim \mathscr N(m_k,P_k)$ with $K_k, m_k, P_k$ exactly as stated. $\blacksquare$

## 5. $\mathbb E[x_k^- \mid y_k^- = y_k^{\mathrm{obs}}]$ and $\mathrm{Var}(x_k^- \mid y_k^- = y_k^{\mathrm{obs}})$

These are precisely the conditional mean and covariance computed in Part 4, since $x_k^+$ was *defined* as that conditional random variable:
$$
\mathbb E[x_k^- \mid y_k^- = y_k^{\mathrm{obs}}] = m_k = m_k^- + K_k(y_k^{\rm obs}-H_km_k^-),
$$
$$
\mathrm{Var}(x_k^- \mid y_k^- = y_k^{\mathrm{obs}}) = P_k = (I-K_kH_k)P_k^-,
$$
with $K_k = P_k^-H_k^T(H_kP_k^-H_k^T+\Sigma_m)^{-1}$ as before.

# Q. 1-D Example

## 1. Prediction step ($a$ scalar)

With $x_k^- = a\,x_{k-1}^+ + w_{k-1}$, $w_{k-1}\sim\mathscr N(0,q)$, this is the scalar case of Part 1 of Question 1 ($A_{k-1}=a$, $G_{k-1}=1$, $\Sigma_p=q$):
$$
m_k^- = a\,m_{k-1}, \qquad P_k^- = a^2P_{k-1} + q. \qquad \blacksquare
$$

## 2. Update step ($h$ scalar)

This is the scalar case of Part 4 of Question 1 with $H_k=h$, $\Sigma_m = r$:
$$
S_k \triangleq h^2P_k^- + r, \qquad K_k = \frac{P_k^-h}{S_k}.
$$
$$
m_k = m_k^- + K_k\,v_k = m_k^- + \frac{P_k^-h}{S_k}\big(y_k^{\rm obs}-hm_k^-\big),
$$
$$
P_k = (1-K_kh)P_k^- = \Big(1-\frac{P_k^-h^2}{S_k}\Big)P_k^-. \qquad \blacksquare
$$

## 3. Predictive measurement distribution $p(y_k^-\mid Y_{k-1})$

From Part 1, $x_k^-\mid Y_{k-1}\sim\mathscr N(m_k^-,P_k^-)$ (this is exactly the prior used in the recursion, conditioned on data up to $k-1$). Then $y_k^- = hx_k^- + z_k$ with $z_k\sim\mathscr N(0,r)$ independent of $x_k^-$, so by the same affine-Gaussian argument as Question 1, Part 2:
$$
p(y_k^-\mid Y_{k-1}) = \mathscr N\big(hm_k^-,\; h^2P_k^- + r\big). \qquad \blacksquare
$$

## 4. Posterior-predictive measurement distribution $p(y_k^-\mid Y_k)$

After filtering, the state distribution conditioned on $Y_k = \{Y_{k-1}, y_k^{\rm obs}\}$ is $x_k^+ \sim \mathscr N(m_k,P_k)$ from Part 2. The *same* measurement model $y_k^- = hx_k^- + z_k$ (with a fresh independent noise realization $z_k$) applied to this updated state gives, again by the affine-Gaussian propagation argument:
$$
p(y_k^-\mid Y_k) = \mathscr N\big(hm_k,\; h^2P_k+r\big). \qquad \blacksquare
$$

## 5. Numerical example and animation

We pick $a=0.95$ (slightly damped scalar dynamics), $q=0.05$ (process noise variance), $h=1$ (direct measurement), $r=0.5$ (measurement noise variance), and animate, at each time step $k$, the **prior predictive** $\mathscr N(m_k^-, P_k^-)$ next to the **posterior** $\mathscr N(m_k,P_k)$ after assimilating the observation $y_k^{\rm obs}$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.stats import norm
from IPython.display import HTML

# ---- model parameters ----
np.random.seed(0)
a, q, h, r = 0.95, 0.05, 1.0, 0.5
m0, P0 = 0.0, 1.0
N = 30

# ---- simulate true state and noisy measurements ----
x_true = np.zeros(N)
y_obs = np.zeros(N)
x_true[0] = m0 + np.sqrt(P0)*np.random.randn()
y_obs[0] = h*x_true[0] + np.sqrt(r)*np.random.randn()
for k in range(1, N):
    x_true[k] = a*x_true[k-1] + np.sqrt(q)*np.random.randn()
    y_obs[k] = h*x_true[k] + np.sqrt(r)*np.random.randn()

# ---- run the 1-D Kalman filter, storing prior and posterior at each step ----
m_prior = np.zeros(N); P_prior = np.zeros(N)
m_post = np.zeros(N); P_post = np.zeros(N)

m_prev, P_prev = m0, P0
for k in range(N):
    # predict
    m_minus = a*m_prev
    P_minus = a**2*P_prev + q
    m_prior[k], P_prior[k] = m_minus, P_minus

    # update
    S = h**2*P_minus + r
    K = P_minus*h/S
    m_k = m_minus + K*(y_obs[k] - h*m_minus)
    P_k = (1 - K*h)*P_minus
    m_post[k], P_post[k] = m_k, P_k

    m_prev, P_prev = m_k, P_k

print("Filter complete. Final estimate: m =", round(m_post[-1],3), " P =", round(P_post[-1],4))
```

```python
# ---- animate prior vs posterior Gaussian at each time step ----
fig, ax = plt.subplots(figsize=(7,4.5))

def frame(k):
    ax.clear()
    xs = np.linspace(-4, 4, 400)
    ax.plot(xs, norm.pdf(xs, m_prior[k], np.sqrt(P_prior[k])), 'b--', lw=2,
            label=f'prior $\\mathcal{{N}}(m_k^-,P_k^-)$')
    ax.plot(xs, norm.pdf(xs, m_post[k], np.sqrt(P_post[k])), 'g-', lw=2,
            label=f'posterior $\\mathcal{{N}}(m_k,P_k)$')
    ax.axvline(x_true[k], color='k', ls=':', label='true state $x_k$')
    ax.axvline(y_obs[k], color='r', ls=':', label='observation $y_k^{obs}$')
    ax.set_title(f'1-D Kalman filter — step k={k}')
    ax.set_xlabel('x'); ax.set_ylabel('density')
    ax.set_xlim(-4,4); ax.set_ylim(0, 1.8)
    ax.legend(loc='upper right', fontsize=8)

anim = FuncAnimation(fig, frame, frames=N, interval=400)
plt.close(fig)
HTML(anim.to_jshtml())
```

# Q. 2D-Position Estimation

## Part A

We discretize the constant-velocity model with sample time $\Delta t$, using the kinematic relations (to first/second order in $\Delta t$, treating the unmodeled acceleration as the process-noise input):
$$
p_x(k) = p_x(k-1) + \Delta t\,v_x(k-1) + \tfrac12\Delta t^2\,u_x, \qquad
v_x(k) = v_x(k-1) + \Delta t\,u_x,
$$
and identically for the $y$-coordinate, where $u_x,u_y$ are the (zero-mean, white) random accelerations driving the process noise $w_{k-1}=\begin{bmatrix}u_x\\u_y\end{bmatrix}\sim\mathscr N(0,\Sigma_p)$.

Writing this in matrix form for the state $x_k=[p_x,p_y,v_x,v_y]^T$:
$$
\underbrace{\begin{bmatrix}p_x(k)\\p_y(k)\\v_x(k)\\v_y(k)\end{bmatrix}}_{x_k^-}
=
\underbrace{\begin{bmatrix}1&0&\Delta t&0\\0&1&0&\Delta t\\0&0&1&0\\0&0&0&1\end{bmatrix}}_{A}
\underbrace{\begin{bmatrix}p_x(k-1)\\p_y(k-1)\\v_x(k-1)\\v_y(k-1)\end{bmatrix}}_{x_{k-1}^+}
+
\underbrace{\begin{bmatrix}\tfrac12\Delta t^2&0\\0&\tfrac12\Delta t^2\\\Delta t&0\\0&\Delta t\end{bmatrix}}_{G}
\underbrace{\begin{bmatrix}u_x\\u_y\end{bmatrix}}_{w_{k-1}},
$$
which reproduces exactly the stated $A$ and $G$. $\blacksquare$

Since only the position components are measured,
$$
y_k = \begin{bmatrix}p_x^{\rm meas}(k)\\p_y^{\rm meas}(k)\end{bmatrix} = \begin{bmatrix}1&0&0&0\\0&1&0&0\end{bmatrix}x_k^+ + z_k = Hx_k^+ + z_k,
$$
which gives exactly the stated selection matrix $H$. $\blacksquare$

## Part B

We implement a general linear Kalman filter class and apply it to a simulated sequence of noisy GPS position measurements generated from a constant-velocity 2-D trajectory using the $A,G,H$ matrices derived in Part A.

```python
import numpy as np

class KalmanFilter:
    \"\"\"Standard linear-Gaussian Kalman filter.

    Model:
        x_k^- = A x_{k-1}^+ + G w_{k-1},   w_{k-1} ~ N(0, Sigma_p)
        y_k   = H x_k^+ + z_k,              z_k     ~ N(0, Sigma_m)
    \"\"\"
    def __init__(self, A, G, H, Sigma_p, Sigma_m, m0, P0):
        self.A, self.G, self.H = A, G, H
        self.Sigma_p, self.Sigma_m = Sigma_p, Sigma_m
        self.m, self.P = m0.copy(), P0.copy()

    def predict(self):
        self.m = self.A @ self.m
        self.P = self.A @ self.P @ self.A.T + self.G @ self.Sigma_p @ self.G.T
        return self.m, self.P

    def update(self, y_obs):
        S = self.H @ self.P @ self.H.T + self.Sigma_m
        K = self.P @ self.H.T @ np.linalg.inv(S)
        self.m = self.m + K @ (y_obs - self.H @ self.m)
        I = np.eye(self.P.shape[0])
        self.P = (I - K @ self.H) @ self.P
        return self.m, self.P

    def step(self, y_obs):
        self.predict()
        return self.update(y_obs)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

# ---- discretization & model matrices (Part A) ----
dt = 0.1
A = np.array([[1,0,dt,0],
              [0,1,0,dt],
              [0,0,1,0],
              [0,0,0,1]])
G = np.array([[0.5*dt**2, 0],
              [0, 0.5*dt**2],
              [dt, 0],
              [0, dt]])
H = np.array([[1,0,0,0],
              [0,1,0,0]])

sigma_accel = 0.5     # std dev of random accelerations driving the process
sigma_gps   = 1.0      # std dev of GPS measurement noise (meters)
Sigma_p = (sigma_accel**2) * np.eye(2)
Sigma_m = (sigma_gps**2) * np.eye(2)

# ---- simulate a true constant-velocity-ish trajectory + noisy GPS fixes ----
N = 150
x_true = np.zeros((N,4))
x_true[0] = [0,0,2.0,1.0]   # start at origin, velocity (2,1) m/s
for k in range(1,N):
    w = sigma_accel*np.random.randn(2)
    x_true[k] = A @ x_true[k-1] + G @ w

y_gps = (H @ x_true.T).T + sigma_gps*np.random.randn(N,2)

# ---- run the Kalman filter on the noisy GPS stream ----
m0 = np.array([y_gps[0,0], y_gps[0,1], 0.0, 0.0])
P0 = np.diag([sigma_gps**2, sigma_gps**2, 4.0, 4.0])

kf = KalmanFilter(A, G, H, Sigma_p, Sigma_m, m0, P0)
estimates = np.zeros((N,4))
estimates[0] = m0
for k in range(1,N):
    m_k, P_k = kf.step(y_gps[k])
    estimates[k] = m_k

# ---- plot results ----
plt.figure(figsize=(7,7))
plt.plot(x_true[:,0], x_true[:,1], 'k-', lw=2, label='true trajectory')
plt.scatter(y_gps[:,0], y_gps[:,1], s=12, c='r', alpha=0.5, label='noisy GPS measurements')
plt.plot(estimates[:,0], estimates[:,1], 'b-', lw=2, label='Kalman filter estimate')
plt.xlabel('x position (m)'); plt.ylabel('y position (m)')
plt.title('Kalman filtering of noisy GPS position measurements')
plt.legend(); plt.axis('equal'); plt.grid(alpha=0.3)
plt.show()

rmse_raw = np.sqrt(np.mean(np.sum((y_gps - x_true[:,:2])**2, axis=1)))
rmse_kf  = np.sqrt(np.mean(np.sum((estimates[:,:2] - x_true[:,:2])**2, axis=1)))
print(f'RMSE of raw GPS measurements:   {rmse_raw:.3f} m')
print(f'RMSE of Kalman filter estimate: {rmse_kf:.3f} m')
```

The Kalman filter estimate tracks the true trajectory much more closely than the raw noisy GPS fixes (lower RMSE), since it fuses the constant-velocity motion model with the noisy position measurements, also recovering smooth velocity estimates that are not directly measured.
