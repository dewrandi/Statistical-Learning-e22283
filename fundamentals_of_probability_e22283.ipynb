### Q1

Let the disk be  
$$
\Omega = \{(x,y) \in \mathbb{R}^2 : x^2 + y^2 \le 10^2\}.
$$

Let \(A\) be the event that the selected point lies within 1 unit of the boundary.  
This means the distance from the center is between 9 and 10:

$$
A = \{(x,y) \in \Omega : 9 \le \sqrt{x^2 + y^2} \le 10\}.
$$

So, \(A\) represents the outer annular region.

---

### (1) Uniform distribution over area

Total area of the disk:
$$
\pi (10)^2 = 100\pi
$$

Area of the annulus:
$$
100\pi - 81\pi = 19\pi
$$

Therefore,
$$
P(A) = \frac{19\pi}{100\pi} = \frac{19}{100}.
$$

---

### (2) Uniform distribution of radius

Radius \(r \sim U(0,10)\)

Total length of possible values:
$$
10
$$

Favorable interval:
$$
[9,10] \Rightarrow \text{length} = 1
$$

Thus,
$$
P(A) = \frac{1}{10}.
$$

---

### (3) Key observation

The results differ because the meaning of randomness changes:

- Area-based model → probability depends on region size  
- Radius-based model → all radii are equally likely  

Hence,
$$
P(A) = \frac{19}{100} \quad \text{and} \quad P(A) = \frac{1}{10}.
$$

### Q2 – Bertrand’s Paradox

Consider the unit circle:
$$
x^2 + y^2 = 1
$$

A chord is longer than the side of the inscribed equilateral triangle (length \( \sqrt{3} \)) **iff** its distance from the center satisfies:
$$
r < \frac{1}{2}
$$

---

### Model 1: Random endpoints (angle-based)

Let the central angle be \( \theta \in [0,\pi] \), uniformly distributed.

Chord length:
$$
L(\theta) = 2\sin\left(\frac{\theta}{2}\right)
$$

Condition:
$$
2\sin\left(\frac{\theta}{2}\right) > \sqrt{3}
\Rightarrow \theta > \frac{2\pi}{3}
$$

Probability:
$$
P = \frac{\pi - \frac{2\pi}{3}}{\pi} = \frac{1}{3}
$$

---

### Model 2: Random distance from center

Let \( r \sim U(0,1) \)

Condition:
$$
r < \frac{1}{2}
$$

Probability:
$$
P = \frac{1}{2}
$$

---

### Model 3: Random midpoint in the disk

Midpoint is uniformly distributed over the unit disk.

Condition:
$$
x^2 + y^2 < \frac{1}{4}
$$

This is a smaller disk of radius \( \frac{1}{2} \)

Probability:
$$
P = \frac{\pi (1/2)^2}{\pi} = \frac{1}{4}
$$

---

### Final Result

$$
P = \frac{1}{3}, \quad \frac{1}{2}, \quad \frac{1}{4}
$$

Different answers arise because each model defines “random chord” differently.

### Q8 – Probability of Two Girls

The sample space for two children is:
$$
\Omega = \{BB, BG, GB, GG\}
$$

Each outcome has probability:
$$
\frac{1}{4}
$$

---

### Given condition

“At least one child is a girl” → exclude \(BB\)

So the reduced sample space becomes:
$$
\{BG, GB, GG\}
$$

---

### Required probability

Only one outcome satisfies both being girls (\(GG\)).

Thus,
$$
P(\text{both girls} \mid \text{at least one girl}) = \frac{1}{3}
$$

---

### Final Answer
$$
\frac{1}{3}
$$

---

### Note

If instead we were told a **specific child** (e.g., older child) is a girl, then:

Possible outcomes:
$$
\{GG, GB\}
$$

So,
$$
P(\text{both girls} \mid \text{older child is a girl}) = \frac{1}{2}
$$

# Q9 — Discrete Random Variables

---

## 1. Probability Space

$$
\Omega=\{G_1,G_2,D_1,D_2\}, \quad \mathfrak{F}=\mathfrak{P}(\Omega)
$$

Probability measure:

$$
P(G_1)=0.50,\quad P(G_2)=0.20,\quad P(D_1)=0.10,\quad P(D_2)=0.20
$$

Check:

$$
0.50+0.20+0.10+0.20=1
$$

---

## 2. Random Variable X (Quality Only)

$$
X(G_1)=0,\quad X(G_2)=0,\quad X(D_1)=1,\quad X(D_2)=1
$$

Preimages:

$$
X^{-1}(0)=\{G_1,G_2\}, \quad X^{-1}(1)=\{D_1,D_2\}
$$

So:

$$
\sigma(X)=\{\varnothing,\Omega,\{G_1,G_2\},\{D_1,D_2\}\}
$$

---

## 3. Random Variable Y (Full Information)

$$
Y(G_1)=1,\quad Y(G_2)=2,\quad Y(D_1)=3,\quad Y(D_2)=4
$$

Since all outcomes are distinct:

$$
\sigma(Y)=\mathfrak{P}(\Omega)
$$

---

## 4. Information Comparison

$$
\sigma(X)\subseteq \sigma(Y)
$$

So:
- $X$ = quality only  
- $Y$ = quality + machine  

---

## 5. Distribution of X

$$
P(X=0)=P(G_1)+P(G_2)=0.70
$$

$$
P(X=1)=P(D_1)+P(D_2)=0.30
$$

---

## 6. Distribution of Y

$$
P(Y=1)=0.50,\quad P(Y=2)=0.20,\quad P(Y=3)=0.10,\quad P(Y=4)=0.20
$$

---

## 7. Joint Probabilities

$$
P(X=0,Y=1)=0.50,\quad P(X=0,Y=2)=0.20
$$

$$
P(X=1,Y=3)=0.10,\quad P(X=1,Y=4)=0.20
$$

---

## 8. Conditional Probabilities

$$
P(X=1\mid Y=3)=1,\quad P(X=1\mid Y=4)=1
$$

$$
P(Y=3\mid X=1)=\frac{0.10}{0.30}=\frac{1}{3}
$$

$$
P(Y=4\mid X=1)=\frac{0.20}{0.30}=\frac{2}{3}
$$

---

## 9. Information Interpretation

- Knowing $Y$ gives exact outcome → no uncertainty  
- Knowing $X$ leaves ambiguity between machines  

So:
- $Y$ gives full information  
- $X$ gives partial information  

---

## 10. Key Conclusion

$$
\sigma(X)\subseteq \sigma(Y)
$$

Thus:

> $Y$ contains more information than $X$, since it refines the partition of the sample space and reduces uncertainty further than $X$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample space probabilities
omega = ["G1", "G2", "D1", "D2"]
p = np.array([0.50, 0.20, 0.10, 0.20])

# Random variables
X = {"G1": 0, "G2": 0, "D1": 1, "D2": 1}
Y = {"G1": 1, "G2": 2, "D1": 3, "D2": 4}

# Marginals
px = {
    0: p[0] + p[1],
    1: p[2] + p[3]
}

py = {
    1: p[0],
    2: p[1],
    3: p[2],
    4: p[3]
}



# Joint distribution table for (X,Y)
# Rows correspond to X = 0, 1
# Columns correspond to Y = 1, 2, 3, 4

joint = np.array([
    [0.50, 0.20, 0.00, 0.00],  # X = 0
    [0.00, 0.00, 0.10, 0.20]   # X = 1
])

fig, ax = plt.subplots(figsize=(8, 4))

im = ax.imshow(joint, cmap="YlOrRd", aspect="auto")

# Axis labels
ax.set_xticks(range(4))
ax.set_xticklabels([1, 2, 3, 4])
ax.set_yticks(range(2))
ax.set_yticklabels([0, 1])

ax.set_xlabel("Y")
ax.set_ylabel("X")
ax.set_title("Heat Map of the Joint Distribution of $(X,Y)$")

# Add probability values inside cells
for i in range(joint.shape[0]):
    for j in range(joint.shape[1]):
        ax.text(j, i, f"{joint[i, j]:.2f}",
                ha="center", va="center", color="black", fontsize=12)

# Color bar
cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Probability")

plt.tight_layout()
plt.show()

# Conditionals
p_y3_given_x1 = p[2] / (p[2] + p[3])
p_y4_given_x1 = p[3] / (p[2] + p[3])

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Visualizing Information in Two Random Variables", fontsize=14)

# 1. Sample space probabilities
axes[0, 0].bar(omega, p, color=["steelblue", "steelblue", "salmon", "salmon"])
axes[0, 0].set_title("Sample space probabilities")
axes[0, 0].set_ylabel("Probability")
for i, val in enumerate(p):
    axes[0, 0].text(i, val + 0.02, f"{val:.2f}", ha="center")

# 2. Partition induced by X
axes[0, 1].axis("off")
axes[0, 1].set_title(r"Partition induced by $\sigma(X)$")
axes[0, 1].text(
    0.1, 0.65,
    "{G1, G2}   ->   X = 0  (Good)\n\n{D1, D2}   ->   X = 1  (Defective)",
    fontsize=12,
    bbox=dict(facecolor="lightyellow", edgecolor="black")
)

# 3. Partition induced by Y
axes[1, 0].axis("off")
axes[1, 0].set_title(r"Partition induced by $\sigma(Y)$")
axes[1, 0].text(
    0.1, 0.45,
    "{G1} -> Y = 1\n{G2} -> Y = 2\n{D1} -> Y = 3\n{D2} -> Y = 4",
    fontsize=12,
    bbox=dict(facecolor="lavender", edgecolor="black")
)

# 4. Marginal of X and conditional of Y given X=1
x_labels = ["P(X=0)", "P(X=1)", "P(Y=3 | X=1)", "P(Y=4 | X=1)"]
x_vals = [px[0], px[1], p_y3_given_x1, p_y4_given_x1]
axes[1, 1].bar(x_labels, x_vals, color=["steelblue", "salmon", "orange", "orange"])
axes[1, 1].set_title("Marginal and conditional probabilities")
axes[1, 1].set_ylabel("Probability")
for i, val in enumerate(x_vals):
    axes[1, 1].text(i, val + 0.02, f"{val:.2f}", ha="center")

plt.tight_layout()
plt.show()
```

#  Q10 — Continuous Random Variables

---

## 1. Probability Space

$$
\Omega=\mathbb{R}^2,\quad \mathfrak{F}=\mathfrak{B}(\mathbb{R}^2)
$$

A point is:

$$
\omega=(t,p)
$$

Joint density:

$$
f_{T,P}(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right)
$$

So:

$$
P(A)=\iint_A f_{T,P}(t,p)\,dt\,dp
$$

with:

$$
T\sim N(0,1),\quad P\sim N(0,4)
$$

---

## 2. Random Variables

$$
X(t,p)=t,\quad Y(t,p)=(t,p)
$$

---

## 3. σ-Algebras

### σ(X)

$$
\sigma(X)=\{B\times\mathbb{R}:B\in\mathfrak{B}(\mathbb{R})\}
$$

---

### σ(Y)

$$
\sigma(Y)=\mathfrak{B}(\mathbb{R}^2)
$$

---

### Relation

$$
\sigma(X)\subset \sigma(Y)
$$

Strict since rectangles like $[-1,1]\times[-2,2]$ are not in $\sigma(X)$.

---

## 4. Marginal of X

$$
f_X(x)=\int_{-\infty}^{\infty} f_{T,P}(x,p)\,dp
$$

Result:

$$
f_X(x)=\frac{1}{\sqrt{2\pi}}e^{-x^2/2}
$$

So:

$$
X\sim N(0,1)
$$

---

## 5. Marginal of Y

$$
f_Y(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right)
$$

---

## 6. Key Probabilities

$$
P(X\le 0)=\frac{1}{2},\quad P(X>1)=1-\Phi(1)\approx 0.1587
$$

$$
P(Y\in(-\infty,0]\times\mathbb{R})=\frac{1}{2}
$$

$$
P(Y\in[-1,1]\times[-2,2])=(2\Phi(1)-1)^2\approx 0.466
$$

---

## 7. Conditional Distributions

### X | Y

$$
X|Y=(t,p)=t \;\Rightarrow\; \delta_t
$$

---

### Y | X

$$
Y|X=t=(t,P),\quad P\sim N(0,4)
$$

---

## 8. Measurability

$$
X=\pi_1\circ Y,\quad \pi_1(t,p)=t
$$

So:

$$
\sigma(X)\subseteq\sigma(Y)
$$

---

## 9. Entropy

### X

$$
h(X)=\frac{1}{2}\ln(2\pi e)\approx 1.4189
$$

---

### Y

$$
h(Y)=h(T)+h(P)
$$

$$
=\frac{1}{2}\ln(2\pi e)+\frac{1}{2}\ln(8\pi e)
$$

$$
\approx 3.5310
$$

---

## 10. Conditional Entropy

$$
h(Y|X)=h(P)=\frac{1}{2}\ln(8\pi e)\approx 2.1121
$$

---

## 11. Key Insight

- Knowing $X$ → temperature only  
- Knowing $Y$ → full system state  

So:

$$
\sigma(X)\subset \sigma(Y)
$$

and:

- $h(Y|X)$ is finite  
- $h(X|Y)$ behaves like $-\infty$ (continuous case pathology)

---

##  Final Conclusion

$$
\sigma(X)\subseteq \sigma(Y)
$$

$Y$ contains strictly more information than $X$, and entropy confirms that conditioning on $Y$ removes all uncertainty about $X$, while conditioning on $X$ still leaves uncertainty in $Y$.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-4, 4, 200)
p = np.linspace(-6, 6, 200)
T, P = np.meshgrid(t, p)

Z = (1/(4*np.pi)) * np.exp(-T**2/2 - P**2/8)

plt.figure()
plt.contourf(T, P, Z)
plt.title("Joint Density of (T,P)")
plt.xlabel("T")
plt.ylabel("P")
plt.show()
```

## 9. Conditioning and Information Interpretation

Conditioning on $X$ means we know only the temperature deviation. The pressure deviation remains uncertain.

Conditioning on $Y$ means we know both temperature and pressure, so it provides strictly more information.

In particular:
- given $X=t$, pressure is still random  
- given $Y=(t,p)$, both values are known exactly  

---

## 10. Measurability

Define the projection map:

$$
\pi_1:\mathbb{R}^2 \to \mathbb{R}, \quad \pi_1(t,p)=t
$$

Then:

$$
X = \pi_1 \circ Y
$$

Indeed:

$$
(\pi_1 \circ Y)(t,p)=\pi_1(t,p)=t=X(t,p)
$$

Since $\pi_1$ is Borel measurable, $X$ is a measurable function of $Y$, hence:

$$
\sigma(X)\subseteq\sigma(Y)
$$

---


A good visualization includes:
- Contour plot of joint density:
$$
f_Y(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right)
$$

- Marginal density:
$$
f_X(x)=\frac{1}{\sqrt{2\pi}}e^{-x^2/2}
$$

- Vertical strips ($B\times \mathbb{R}$) → events in $\sigma(X)$  
- Rectangles like $[-1,1]\times[-2,2]$ → events in $\sigma(Y)$  

 Key idea:  
$\sigma(X)$ sees only vertical strips, while $\sigma(Y)$ sees all Borel sets in $\mathbb{R}^2$.

## 11.Entropy Results

Since $X \sim N(0,1)$:

$$
h(X)=\frac{1}{2}\ln(2\pi e) \approx 1.4189
$$

---

Since $Y=(T,P)$ with independent components:

$$
h(Y)=h(T)+h(P)
$$

$$
h(T)=\frac{1}{2}\ln(2\pi e), \quad
h(P)=\frac{1}{2}\ln(8\pi e)
$$

So:

$$
h(Y)=\frac{1}{2}\ln(2\pi e)+\frac{1}{2}\ln(8\pi e)
$$

$$
h(Y)\approx 3.5310
$$

---

## 12. Conditional Entropy

Since $T$ and $P$ are independent:

$$
h(Y|X)=h(P)=\frac{1}{2}\ln(8\pi e)\approx 2.1121
$$

Interpretation: once temperature is known, only pressure uncertainty remains.

---

## 13. Deterministic Dependence

Since $X$ is fully determined by $Y$, one might expect:

$$
h(X|Y)=0
$$

However, in continuous distributions:

$$
h(X|Y)\to -\infty
$$

This is due to the conditional distribution being degenerate, a key difference from discrete entropy.

---

## 📌 Key Summary

- $\sigma(X)\subseteq\sigma(Y)$  
- $Y$ contains more information than $X$  
- $h(Y|X)$ is meaningful and positive  
- $h(X|Y)$ behaves differently in continuous settings compared to discrete cases

## 16. Why $h(X|Y)\neq 0$ (Continuous Case)

Since $X$ is a deterministic function of $Y$:

$$
X = \pi_1(Y)
$$

one might expect:

$$
h(X|Y)=0
$$

However, in continuous distributions:

- the conditional distribution of $X|Y$ is degenerate
- differential entropy is not well-behaved under such determinism

So:

$$
h(X|Y)\to -\infty
$$

---

### Key difference

- **Discrete case:** if $X$ is determined by $Y$, then $H(X|Y)=0$
- **Continuous case:** deterministic dependence leads to singular distributions and entropy behaves differently
