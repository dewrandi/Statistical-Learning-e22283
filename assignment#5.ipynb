## Part A — Sample Answer


1. Derivation of the Bias in the MLE Covariance Matrix

We begin with the Maximum Likelihood Estimator for the population covariance matrix $\boldsymbol{\Sigma}$:


$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$


where $\widehat{\boldsymbol{\mu}}_n = \frac{1}{n}\sum_{i=1}^n \mathbf{X}_i$ is the empirical sample mean.

To determine if this estimator is biased, we expand the deviations around the true, hidden population mean $\boldsymbol{\mu}$:


$$\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n = (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$$

Substituting this back into the sum inside the MLE formula gives:


$$\sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T = \sum_{i=1}^n \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right]^T$$

Expanding the product matrix term-by-term yields:


$$= \sum_{i=1}^n (\mathbf{x}_i - \boldsymbol{\mu})(\mathbf{x}_i - \boldsymbol{\mu})^T - \sum_{i=1}^n (\mathbf{x}_i - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T - \sum_{i=1}^n (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\mathbf{x}_i - \boldsymbol{\mu})^T + \sum_{i=1}^n (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

Notice that $\sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu}) = n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$. Using this simplification, the two middle cross-product terms contract:


$$= \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T + n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

$$= \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

Now, we apply the expectation operator $\mathbb{E}[\cdot]$ to both sides:


$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} \sum_{i=1}^n \mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T] - \mathbb{E}[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T]$$

By definition of the structural population moments:

* $\mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T] = \boldsymbol{\Sigma}$
* $\mathbb{E}[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T] = \text{Var}(\widehat{\boldsymbol{\mu}}_n) = \frac{1}{n}\boldsymbol{\Sigma}$

Substituting these back into our expectation equation gives:


$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} (n\boldsymbol{\Sigma}) - \frac{1}{n}\boldsymbol{\Sigma} = \boldsymbol{\Sigma} - \frac{1}{n}\boldsymbol{\Sigma} = \left(\frac{n-1}{n}\right)\boldsymbol{\Sigma}$$

Because $\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] \neq \boldsymbol{\Sigma}$, the estimator is systematically downward biased. This occurs because calculating variations against the *sample* mean rather than the *true* population mean artificially underrepresents the real spatial scattering of the system.

**Bessel's Correction Elimination of Bias**

To yield an unbiased estimator ($\mathbf{S}$), we scale the maximum likelihood matrix by the reciprocal of the bias factor ($\frac{n}{n-1}$):


$$\mathbf{S} = \frac{n}{n-1} \widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{n}{n-1} \left( \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T \right) = \frac{1}{n-1} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

Evaluating the expectation confirms its lack of bias:


$$\mathbb{E}[\mathbf{S}] = \frac{n}{n-1} \mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{n}{n-1} \left(\frac{n-1}{n}\right)\boldsymbol{\Sigma} = \boldsymbol{\Sigma}$$

---

**Operational Error Diagnostics**

* **Type I Error ($\alpha$):** A **False Alarm**. This happens when the diagnostic system concludes that the structure is damaged or unstable (rejects $H_0$), when in reality the physical system remains completely nominal and healthy.
* **Type II Error ($\beta$):** A **Missed Detection**. This occurs when a structural shift or localized crack has compromised the asset (rejects $H_1$), but the diagnostic system fails to alert the team (fails to reject $H_0$).

**Consequence of an Ultra-Conservative Threshold ($\alpha = 0.0001$)**

When $\alpha$ is reduced dramatically to eliminate false alarms, the critical value boundary of the test statistic is pushed deep into the extreme tails of the baseline distribution.

1. **Impact on Statistical Power ($1-\beta$):** Because the threshold for detecting anomalies becomes incredibly stringent, minor or emerging structural shifts will fail to cross the new partition line. Therefore, $\beta$ (missed detections) will sharply escalate, causing the **statistical power ($1-\beta$) of the model to collapse**.
2. **Impact on Confidence Ellipsoids:** The confidence level of the multi-dimensional boundary scales as $1-\alpha = 0.9999$. This means the **geometric volume of the healthy operational ellipsoid expands significantly**. The boundary swells to encompass almost all variation, making the system highly insensitive to genuine baseline degradation.

## Part B — Sample Answer


Let $\{X_n\}$ and $\{Y_n\}$ be sequences of random variables (or vectors). If $X_n$ converges in distribution to a random variable $X$ ($X_n \xrightarrow{d} X$), and $Y_n$ converges in probability to a constant scalar or matrix $c$ ($Y_n \xrightarrow{p} c$), then:

* $X_n + Y_n \xrightarrow{d} X + c$
* $X_n Y_n \xrightarrow{d} cX$
* $X_n / Y_n \xrightarrow{d} X/c$ (provided $c \neq 0$)

**Operational Distribution Justification**

The Lindeberg-Lévy Multivariate CLT asserts that:


$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})$$

Multiplying by the inverse matrix square root of the population covariance normalizes the space:


$$\boldsymbol{\Sigma}^{-1/2} \left[ \sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \xrightarrow{d} \mathscr{N}(\mathbf{0}, \mathbf{I})$$

By the Weak Law of Large Numbers (WLLN), the unbiased empirical sample covariance matrix $\mathbf{S}$ is a continuous, consistent estimator of the population matrix $\boldsymbol{\Sigma}$. Therefore, as $n \to \infty$, $\mathbf{S}$ converges in probability to the true constant profile:


$$\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma} \implies \mathbf{S}^{-1/2} \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}$$

Applying **Slutsky's Theorem**, we multiply the sequence converging in distribution by the sequence converging in probability to a constant matrix:


$$\mathbf{S}^{-1/2} \left[ \sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \xrightarrow{d} \boldsymbol{\Sigma}^{-1/2} \cdot \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma}) = \mathscr{N}(\mathbf{0}, \mathbf{I})$$

Because the target boundary converges asymptotically to a standard Gaussian space, we scale the localized coordinate parameters back to the empirical matrix metric:


$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \sim \mathscr{N}(\mathbf{0}, \mathbf{S}) \implies \widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\boldsymbol{\mu}, \frac{1}{n}\mathbf{S}\right)$$

This mathematically validates substituting $\mathbf{S}$ for $\boldsymbol{\Sigma}$ without corrupting the stability of the asymptotic diagnostic envelope.

## Part A — Implementation


```python
import numpy as np
import pandas as pd
from typing import Optional, Sequence, Dict, Any
from sklearn.decomposition import FactorAnalysis
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# =====================================================================
# COMBINED SUBSPACE ENGINES (PCA & FA PIPELINES)
# =====================================================================

class DualSubspaceDiagnosticsEngine:
    def __init__(self, df: pd.DataFrame):
        self.df = df

    def compute_empirical_pca(self, columns: Optional[Sequence[str]] = None, show_plot: bool = True) -> Dict[str, Any]:
        """
        Operationalizes the geometric de-correlation framework via PCA.
        Decomposes the unbiased empirical covariance matrix S into its orthogonal
        basis P_hat and renders a clean 2x3 optimization and feature contribution dashboard.
        """
        if self.df is None:
            raise ValueError("Error: No data loaded.")

        if columns is None:
            target_cols = self.df.select_dtypes(include=[np.number]).columns.tolist()
            if 'count' in target_cols: target_cols.remove('count')
            target_cols = list(target_cols)
        else:
            target_cols = list(columns)

        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape

        if n <= m:
            raise ValueError("Sample size n must be larger than feature dimensions m to isolate principal components.")

        # 1. Centering
        mu_hat = np.mean(X, axis=0)
        X_centered = X - mu_hat

        # 2. Compute Unbiased Sample Covariance Matrix S
        S_matrix = np.cov(X, rowvar=False, ddof=1)

        # 3. Spectral Decomposition
        eigenvalues, eigenvectors = np.linalg.eigh(S_matrix)
        idx = np.argsort(eigenvalues)[::-1]
        lambda_hat = eigenvalues[idx]
        P_hat = eigenvectors[:, idx]
        lambda_hat = np.clip(lambda_hat, a_min=1e-15, a_max=None)

        # 4. Variance Information Calculations
        total_variance = np.sum(lambda_hat)
        if total_variance == 0:
            raise ValueError("Total variance is zero.")

        explained_variance_ratio = lambda_hat / total_variance
        cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
        unexplained_variance_ratio = 1.0 - cumulative_variance_ratio

        # 5. Map Realizations to Principal Component Scores
        Z_scores = np.dot(X_centered, P_hat)
        S_Z = np.cov(Z_scores, rowvar=False, ddof=1)

        # 6. Evaluate Statistical Distance Metrics (T^2 and Q)
        k_range = np.arange(1, m)
        mean_T2_vs_k = []
        mean_Q_vs_k = []
        T2_matrix = np.zeros((n, len(k_range)))
        Q_matrix = np.zeros((n, len(k_range)))

        for idx_k, k in enumerate(k_range):
            Z_k = Z_scores[:, :k]
            lambda_k = lambda_hat[:k]
            T2_samples = np.sum((Z_k ** 2) / lambda_k, axis=1)
            T2_matrix[:, idx_k] = T2_samples
            mean_T2_vs_k.append(np.mean(T2_samples))

            Z_residual = Z_scores[:, k:]
            Q_samples = np.sum(Z_residual ** 2, axis=1)
            Q_matrix[:, idx_k] = Q_samples
            mean_Q_vs_k.append(np.mean(Q_samples))

        print(f"\n--- Operationalizing Geometric De-correlation Layer (PCA) ---")
        print(f"Decomposing structural space of {m} features using {n} samples.")
        print(f"Total System Variance (Trace[S]): {total_variance:.4f}")

        # 7. Enhanced 2x3 Diagnostic Dashboard
        if show_plot:
            pc_labels = [f"PC {i+1}" for i in range(m)]
            k_labels = [f"k={k}" for k in k_range]
            sensor_labels = target_cols

            fig = make_subplots(
                rows=2, cols=3,
                horizontal_spacing=0.18,
                vertical_spacing=0.28,
                subplot_titles=(
                    "Feature Loading Matrix |P_hat|",
                    "Component Values (Eigenvalues λ)",
                    "Information Profile (Explained Var.)",
                    "Residual Space (Unexplained Var.)",
                    "Mean Hotelling's T² vs Subspace Size k",
                    "Mean Q Statistic (SPE) vs Subspace Size k"
                )
            )

            # --- ROW 1, COL 1: PCA Loading Heatmap (Left-anchored Colorbar) ---
            fig.add_trace(
                go.Heatmap(
                    z=np.abs(P_hat),
                    x=pc_labels,
                    y=sensor_labels,
                    colorscale='YlOrRd',
                    colorbar=dict(
                        title="Loading Weight",
                        x=-0.12,          # Perfectly isolates colorbar to the left side
                        len=0.38,         # Fits height bounds of row 1
                        y=0.78,           # Centers vertically inside row 1
                        yanchor="middle",
                        xanchor="right",
                        titleside="top"
                    ),
                    showscale=True,
                    showlegend=False
                ),
                row=1, col=1
            )
            fig.update_xaxes(title_text="Principal Axes", row=1, col=1)

            # --- ROW 1, COL 2: Absolute Eigenvalues ---
            fig.add_trace(go.Bar(x=pc_labels, y=lambda_hat, name="Eigenvalue (λ_j)", marker=dict(color='#1f77b4', line=dict(color='black', width=0.5)), legendgroup="eigen"), row=1, col=2)
            fig.update_yaxes(title_text="Variance Magnitude", row=1, col=2)
            fig.update_xaxes(title_text="Principal Axes", row=1, col=2)

            # --- ROW 1, COL 3: Variance Profiles ---
            fig.add_trace(go.Bar(x=pc_labels, y=explained_variance_ratio * 100, name="Marginal Explained", marker=dict(color='#ff7f0e', opacity=0.75), legendgroup="expl"), row=1, col=3)
            fig.add_trace(go.Scatter(x=pc_labels, y=cumulative_variance_ratio * 100, mode='lines+markers', name='Cumulative Captured', line=dict(color='#d62728', width=2.5, dash='dash'), legendgroup="expl"), row=1, col=3)
            fig.update_yaxes(title_text="Captured Structure (%)", range=[-2, 105], row=1, col=3)
            fig.update_xaxes(title_text="Principal Axes", row=1, col=3)

            # --- ROW 2, COL 1: Remaining Noise ---
            fig.add_trace(go.Bar(x=pc_labels, y=unexplained_variance_ratio * 100, name="Remaining Noise", marker=dict(color='#2ca02c', line=dict(color='black', width=0.5)), legendgroup="noise"), row=2, col=1)
            fig.update_yaxes(title_text="Excluded Info (%)", range=[-2, 105], row=2, col=1)
            fig.update_xaxes(title_text="Principal Axes", row=2, col=1)

            # --- ROW 2, COL 2: Hotelling's T^2 ---
            fig.add_trace(go.Scatter(x=k_labels, y=mean_T2_vs_k, mode='lines+markers', name='Mean T²', line=dict(color='#9467bd', width=2.5), marker=dict(size=6, symbol='diamond'), legendgroup="t2"), row=2, col=2)
            fig.update_yaxes(title_text="Average T² Metric", row=2, col=2)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=2)

            # --- ROW 2, COL 3: SPE / Q Statistic ---
            fig.add_trace(go.Scatter(x=k_labels, y=mean_Q_vs_k, mode='lines+markers', name='Mean Q (SPE)', line=dict(color='#e377c2', width=2.5), marker=dict(size=6, symbol='square'), legendgroup="q_stat"), row=2, col=3)
            fig.update_yaxes(title_text="Average Residual Energy", row=2, col=3)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=3)

            # Global Unified Layout Configurations
            fig.update_layout(
                title=dict(text="Principal Component Analysis (PCA) Optimization & Feature Loading Dashboard", x=0.5, y=0.97, xanchor="center", yanchor="top"),
                template="plotly_white",
                showlegend=True,
                legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=1.02,
                    xanchor="center",
                    x=0.5
                ),
                margin=dict(t=150, b=60, l=140, r=80),
                height=750,
                width=1250
            )
            fig.show()

        return {
            "mean_vector": mu_hat, "covariance_matrix_S": S_matrix, "eigenvalues_lambda": lambda_hat,
            "eigenvectors_P": P_hat, "explained_variance_ratio": explained_variance_ratio,
            "cumulative_variance_ratio": cumulative_variance_ratio, "unexplained_variance_ratio": unexplained_variance_ratio,
            "transformed_scores_Z": Z_scores, "score_covariance_diagonal": np.diag(S_Z),
            "features": target_cols, "k_values": k_range, "T2_matrix_vs_k": T2_matrix, "Q_matrix_vs_k": Q_matrix,
            "mean_T2_profile": np.array(mean_T2_vs_k), "mean_Q_profile": np.array(mean_Q_vs_k)
        }

    def compute_empirical_fa(self, k: int, columns: Optional[Sequence[str]] = None, show_plot: bool = True) -> Dict[str, Any]:
        """
        Operationalizes the Factor Analysis latent subspace framework with a clean 2x2 grid.
        Anchors the heatmap colorbar perfectly to the right edge of the first subplot.
        """
        if self.df is None:
            raise ValueError("Error: No data loaded.")

        if columns is None:
            target_cols = self.df.select_dtypes(include=[np.number]).columns.tolist()
            if 'count' in target_cols: target_cols.remove('count')
            target_cols = list(target_cols)
        else:
            target_cols = list(columns)

        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape

        # 1. Standardize realizations to construct Z space
        mu_hat = np.mean(X, axis=0)
        std_hat = np.std(X, axis=0, ddof=1)
        std_hat[std_hat == 0] = 1e-15
        Z = (X - mu_hat) / std_hat
        R_matrix = np.corrcoef(X, rowvar=False)

        # 2. Fit Factor Analysis Model (Varimax Rotation)
        fa = FactorAnalysis(n_components=k, rotation='varimax', random_state=42)
        fa.fit(Z)
        lambda_matrix = fa.components_.T
        psi_diagonal = fa.noise_variance_

        # 3. Calculate Communality (h²) and Uniqueness (φ²) vectors
        communality = np.sum(lambda_matrix**2, axis=1)
        uniqueness = psi_diagonal

        # 4. Map realizations using Thomson's Regression Method
        F_scores = fa.transform(Z)

        print(f"\n--- Operationalizing Latent Subspace Layer (Factor Analysis) ---")
        print(f"Decomposing structural space of {m} sensors into {k} hidden physical factors.")
        print(f"Average System Communality (Shared Subspace Energy): {np.mean(communality)*100:.2f}%")
        print(f"Average System Uniqueness (Localized Noise Floor): {np.mean(uniqueness)*100:.2f}%")

        # 5. Clean 2x2 Dashboard
        if show_plot:
            sensor_labels = target_cols
            factor_labels = [f"Factor {j+1}" for j in range(k)]

            fig = make_subplots(
                rows=2, cols=2, horizontal_spacing=0.24, vertical_spacing=0.28,
                subplot_titles=(
                    "Structural Loadings Matrix |λ_(j,r)|",
                    "Variance Partitioning (Communality vs Uniqueness)",
                    "Sensor Uniqueness Noise Floor (φ²)",
                    "Latent Factor Scores Empirical Variance"
                )
            )

            # --- Subplot 1: Factor Loadings Heatmap (Colorbar repositioned to x=0.42) ---
            fig.add_trace(
                go.Heatmap(
                    z=np.abs(lambda_matrix),
                    x=factor_labels,
                    y=sensor_labels,
                    colorscale='YlOrRd',
                    colorbar=dict(
                        title="Sensitivity",
                        x=-0.1,          # Shifts colorbar out of the far right to center-left
                        len=0.38,         # Scales down length to match row 1 height bounds
                        y=0.78,          # Centers vertically inside row 1
                        yanchor="middle",
                        xanchor="right"
                    ),
                    showscale=True,
                    showlegend=False
                ),
                row=1, col=1
            )
            fig.update_xaxes(title_text="Latent Structures", row=1, col=1)

            # --- Subplot 2: Variance Partitioning ---
            fig.add_trace(go.Bar(y=sensor_labels, x=communality * 100, name="Communality (h² - Shared Structure)", orientation='h', marker=dict(color='#1f77b4'), legendgroup="communality"), row=1, col=2)
            fig.add_trace(go.Bar(y=sensor_labels, x=uniqueness * 100, name="Uniqueness (φ² - Channel Noise)", orientation='h', marker=dict(color='#ff7f0e'), legendgroup="uniqueness_bar"), row=1, col=2)
            fig.update_layout(barmode='stack')
            fig.update_xaxes(title_text="Variance Allocation (%)", range=[0, 100], row=1, col=2)

            # --- Subplot 3: Uniqueness Line Profile ---
            fig.add_trace(go.Scatter(x=sensor_labels, y=uniqueness, mode='lines+markers', name='Uniqueness Profile (φ²)', line=dict(color='#d62728', width=2, dash='dot'), marker=dict(size=8, symbol='x'), legendgroup="uniqueness_line"), row=2, col=1)
            fig.update_yaxes(range=[-0.05, 1.05], row=2, col=1)
            fig.update_xaxes(title_text="Monitored Channels", tickangle=25, row=2, col=1)

            # --- Subplot 4: Latent Factor Scores Empirical Variance ---
            factor_variances = np.var(F_scores, axis=0, ddof=1)
            fig.add_trace(go.Bar(x=factor_labels, y=factor_variances, name="Factor Empirical Variance", marker=dict(color='#2ca02c', line=dict(color='black', width=0.5)), legendgroup="variance"), row=2, col=2)
            fig.update_yaxes(title_text="Variance Level", row=2, col=2)
            fig.update_xaxes(title_text="Latent Vectors", row=2, col=2)

            # Global Layout Updates
            fig.update_layout(
                title=dict(text="Factor Analysis (FA) Latent Subspace Diagnostics Dashboard", x=0.5, y=0.97, xanchor="center", yanchor="top"),
                template="plotly_white",
                showlegend=True,
                legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=1.02,
                    xanchor="center",
                    x=0.5,
                    traceorder="normal",
                    itemwidth=40,
                    itemsizing="constant"
                ),
                margin=dict(t=100, b=60, l=80, r=80),
                height=750,
                width=1100
            )
            fig.show()

        return {
            "mean_vector_mu": mu_hat, "std_vector_D": std_hat, "correlation_matrix_R": R_matrix,
            "factor_loadings_lambda": lambda_matrix, "uniqueness_psi": uniqueness,
            "communality_h2": communality, "latent_factor_scores_F": F_scores, "sensors": target_cols
        }

# =====================================================================
# DATA GENERATOR & PIPELINE EXECUTION
# =====================================================================
if __name__ == "__main__":
    # Set execution seed for perfect reproducible properties
    np.random.seed(42)
    n_samples = 2500

    # 1. Generate underlying, independent true physical driving forces (Latent Factors)
    f1 = np.random.normal(0, 1, n_samples)  # Primary structural mode/load
    f2 = np.random.normal(0, 1, n_samples)  # Secondary operational mode

    # 2. Construct the Observable Multi-Sensor Array with targeted noise profiles
    # Sensors 1 and 2 couple strongly to the primary process (f1)
    s1 = 0.85 * f1 + 0.10 * f2 + np.random.normal(0, 0.3, n_samples)
    s2 = 0.80 * f1 + 0.15 * f2 + np.random.normal(0, 0.35, n_samples)

    # Sensor 3 couples strongly to the secondary process (f2)
    s3 = 0.12 * f1 + 0.90 * f2 + np.random.normal(0, 0.25, n_samples)

    # Sensor 4 suffers an extreme electrical/instrumentation fault (massive localized white noise)
    s4 = 0.02 * f1 + 0.05 * f2 + np.random.normal(0, 1.40, n_samples)

    # 3. Compile into the final evaluation DataFrame
    df_asset = pd.DataFrame(
        data=np.vstack([s1, s2, s3, s4]).T,
        columns=['Sensor_1', 'Sensor_2', 'Sensor_3', 'Sensor_4']
    )
    # Instantiate Engine
    engine = DualSubspaceDiagnosticsEngine(df_asset)

    # Run PCA Engine (1x5 Dashboard)
    pca_metrics = engine.compute_empirical_pca(show_plot=True)

    # Run FA Engine (2x2 Dashboard, Retaining k=2 Common Physical Factors)
    fa_metrics = engine.compute_empirical_fa(k=2, show_plot=True)
```

## Part B — Analysis Answers


### Solution for Question 1: The Total Variance Illusion

* **Part 1 (Physical Nature of Sensor 4):** A Uniqueness value ($\varphi^2$) approaching $100\%$ indicates that `Sensor_4` shares zero covariance with the rest of the array. It is completely decoupled from the real physical processes ($f_1, f_2$) driving the plant asset. Its data profile consists entirely of isolated white noise or severe electrical fault energy.
* **Part 2 (The PCA Loading Contamination):** Looking at the **Feature Loading Matrix** in the PCA dashboard, we see a striking anomaly: `PC 1` maps almost exclusively to a single variable—`Sensor_4` (indicated by the deep red block with a loading value near $1.0$). Because `Sensor_4` has a massive, unmitigated noise variance scale ($\sigma^2 \approx 2.0$), the PCA engine maximizes global variance by dedicating its entire primary axis (`PC 1`) to tracking this single noisy sensor. It completely isolates `Sensor_1`, `Sensor_2`, and `Sensor_3` to a loading of $0$ on `PC 1`.
* **Part 3 (Engineering Risk):** This presents a significant operational hazard. If you use standard PCA for monitoring, a single blown or short-circuited instrument will hijack your most critical monitoring axis (`PC 1`), inflating its eigenvalue to $\lambda_1 \approx 2.0$. This masks the true common underlying structural dynamics of the asset and renders your primary control chart useless for catching actual systemic failures.

### Solution for Question 2: Decoupling Structural Loading via Rotation

* **Part 1 (Mathematical Jumble vs. Simple Structure):** Because PCA is bound to strict, orthogonal variance maximization, look at what happens to the remaining signals on `PC 2` and `PC 4`. `PC 2` and `PC 4` exhibit high loading values across *both* `Sensor_1` and `Sensor_2` simultaneously, while also bleeding into `Sensor_3`. This cross-loading represents a mathematically forced mixture of signals. Factor Analysis applies a **Varimax rotation** to eliminate this. By maximizing loading variance within columns, it completely isolates the signals: `Sensor_1` and `Sensor_2` group cleanly into `Factor 1`, while `Sensor_3` shifts entirely into `Factor 2`.
* **Part 2 (Operational Interpretability):** The PCA loading matrix forces a complex interpretation. If a diagnostic alarm trips on `PC 2`, an engineer has to figure out a messy combination of Sensors 1, 2, and 3. With the rotated FA framework, the layout is intuitive. `Factor 1` represents the isolated physical process causing redundancy in `Sensor_1` and `Sensor_2`, while `Factor 2` maps directly to the independent process monitored by `Sensor_3`. Troubleshooting root causes becomes instantaneous.


### Solution for Question 3: Determining Subspace Truncation ($k$)

* **Part 1 (Curve Observations):** Moving from $k=1$ to $k=2$, the Mean $Q$ Statistic (SPE) plunges from $2.4$ down to $0.9$. When transitioning from $k=2$ to $k=3$, the curve flattens out into an explicit "elbow", decreasing at a much slower, linear rate down toward $0.1$.
* **Part 2 (The True Intrinsic Dimension):** Cross-referencing the $Q$ curve with the new loading matrix makes the selection clear. `PC 1` tracks the isolated noise of `Sensor_4`. `PC 2` captures the true, shared variance of the working sensors (`Sensor_1`, `Sensor_2`, `Sensor_3`). The sharp drop in the $Q$ statistic at $k=2$ tells us that once these two distinct behaviors are accounted for, the structured information is fully exhausted. Thus, $k=2$ represents the true hidden physical dimensionality of the asset.
* **Part 3 (The Danger of Over-fitting to $k=3$):** Look at `PC 3` in the loading matrix—it maps exclusively to the leftover residual variance of `Sensor_3`. Setting your truncation limit to $k=3$ pulls this residual space into your "clean" model subspace. This overfits the system, causing your Hotelling's $T^2$ control limits to trigger frequent false alarms due to minor, normal baseline variations.



### Solution for Question 4: Operational Trade-offs in System Health Monitoring

* **The PCA Strategy:** Tracking the main principal subspace via $T^2$ alongside the residual space via the $Q$ statistic.
* *Vulnerability:* This approach is highly vulnerable to sensor degradation. As shown in the PCA loading matrix, if a single channel (`Sensor_4`) fails high, it structurally shifts the orientation of your principal axes, dragging the calculations for the remaining healthy channels along with it.


* **The FA Strategy:** Directly monitoring the rotated Latent Factor Scores.
* *Robustness:* This approach is highly robust. Because Factor Analysis explicitly separates localized noise via the diagonal uniqueness vector ($\Psi$), any single-channel sensor failure is safely absorbed by that channel's uniqueness profile ($\psi_4^2 \rightarrow 1$). The underlying latent factors (`Factor 1` and `Factor 2`) remain clean, stable, and uncorrupted, allowing the rest of your health monitoring network to function normally.
